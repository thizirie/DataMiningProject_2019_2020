```python
#looking for missing value correlation
import missingno as mn
```

```python
#looking for missing variables in the dataset. The white stripes represent missing values
mn.matrix(training_data)
plt.show()
```

```python
#Nullity correlation ranges from (1 to -1). Variables with a variance of zero (variables are always full or empty )
#have no meaningful correlation and are remove from the correlation visualization
mn.heatmap(training_data)
plt.show()
```

```python
#dendrogram:It uses a hierrachical clustering algorithm to bin variables against one another by their nullity correlation
mn.dendrogram(training_data)
```

```python
# nullity filtering of values in the dataset using matrix
filtered_data = mn.nullity_filter(training_data, filter='bottom', n=15, p=0.999) # or filter='top'
mn.matrix(filtered_data.sample(250))
plt.show()
```

```python
# Nullity correlation for missing values
#Nullity of Correlation is meaningless, it does not give any positive analysis. In our visualization we 
#can properly see that â€œTRIMâ€ and â€œWhellTypeâ€ are completely full and give no meaningful information because of the nullity 
#of correlation that exist in them.
mn.heatmap(training_data)
plt.show()
```

```python
#Correlation of numeric variables with IsBadBuy. Only 10% IsBadBuy
#Correlation values ranges from 1 to -1. The closer the value to 1 or -1 means there are high correlation between attributes. 
#If a value is close to zero or is zero that means there is no relationship between variables. In our heat-map below we notice 
#that the diagonals are all 1, this is because those points are correlating each variable to itself. In the remaining variables,
#the greater the value means there is a high correlation between the two attributes. In the positive correlation coefficients 
#the values of both variables tend to increases whereas in the Negative coefficients one variable value increase the other decreases.

plt.subplots(figsize =(14, 12))
hm = sns.heatmap(training_data.corr(), 
           xticklabels=training_data.corr().columns.values,
           yticklabels=training_data.corr().columns.values, annot=True)
```

```python
#Correlation of variables with VehYear. We know from our previous bar charts that 90% is a Good buy.
#Since we know that only 90% of the dealings are good buy and only 10% is good buy from our initial bar graph plot. We would 
#therefore plot a heat-map that would display only the transactions of the Good buys. The â€œIsBadBuyâ€ attribute would be drop 
#since is not our target variable in this heat-map visualization. Also variables such as â€œBYRNOâ€ and â€œIsOnlineSaleâ€ attributes
#will also be excluded from our visualization since it provides less useful information for our analysis. Values of one (1) 
#represent high correlation because the points correlate each attribute to itself. As for the remaining points the greater the 
#value the higher the correlation it is.
plt.subplots(figsize =(14, 12))
hm = sns.heatmap(training_data.corr(), 
           xticklabels=training_data.corr().columns.values,
           yticklabels=training_data.corr().columns.values, annot=True)
```
